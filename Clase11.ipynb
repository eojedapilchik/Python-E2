## Funciones Importantes de Pandas

```python
import pandas as pd
data = pd.read_csv('MRNA.csv',index_col=0)
```

```python
## Porcentaje de variacion entre filas, con respecto a la fila anterior o la siguiente (puede tener saltos)
data['var_diaria']= data['Adj Close'].pct_change()*100 
# si ponen negativos es las filas siguientes, los positivos son con respecto a las filas anteriores
data
```

```python
data['var_diaria']= data['Adj Close'].pct_change(-10)*100 # porcentaje de cambio con respecto a 
#decima fila siguiente
data
```

```python
## Calcular la diferencia entre dos filas
data['var_precio']=data.Close.diff()
data
```

```python
data['var_precio']=data.Close.diff(5)
data
```

```python
# Evaluar una condicion
data['subida_10%'] = data.eval('Close > Open *1.1')
data
```

```python
# Shift, les permite acceder a los datos de la fila anterior para cualquier columnas
# shift(-1) es la fila siguiente, shift(n) n-enesima fila anterior
```

```python
import pandas as pd
data = pd.read_csv('MRNA.csv',index_col=0)
```

```python
data['cierre_ant']=data.Close.shift()
data['gap_positvo']=data.Open > data['cierre_ant']
data
```
