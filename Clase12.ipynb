## Leyendas, Etiquetas de los ejes, Titulos

```python
# Crear la figura, especificando el tamaÃ±o del canvas
grafico = plt.figure(figsize=(8,4), dpi=100)


axes = grafico.add_axes([0.1,0.1,0.8,0.8]) # Izquierda, parte inferior, ancho, alto (valores del 0 al 1)
axes.set_xlabel(f"Eje X")
axes.set_ylabel(f"Eje Y")
axes.set_title(f"Grafica")
axes.plot(x,x**2,label="y1 = X^2") # si no se le pone label, no sale la leyenda.
axes.plot(x,x**3,label="y2 = X^3")
# Leyenda
axes.legend(loc=1) # del 1 al 4. Se arranca en el orden contrario a las agujas del reloj. Siendo 1 la superior derecha.
axes.legend(loc=2)
axes.legend(loc=3)
axes.legend(loc=4)
axes.legend()
```

# Colores de las lineas, ancho de la linea, tipo de linea

```python
fig, ax = plt.subplots()
ax.plot(x,x**2,'y*--', label="y1 = X^2") # si no se le pone label, no sale la leyenda.
ax.plot(x,x**3,'b--', label="y2 = X^3")
```

```python
# color con codigo hex
fig, ax = plt.subplots()
ax.plot(x,x**2,color="blue", label="y1 = X^2") # si no se le pone label, no sale la leyenda.
ax.plot(x,x**3,color="#fc03fc", label="y1 = X^2")
```

```python
fig, ax = plt.subplots(figsize=(8,4))
# Parametro ancho de linea
ax.plot(x,x+2,color="blue", linewidth=0.50)
ax.plot(x,x+3,color="blue", linewidth=1)
ax.plot(x,x+4,color="blue", lw=2)

# Parametro del tipo de linea
ax.plot(x,x+5,color="green",linestyle="-")# continua
ax.plot(x,x+6,color="green",ls="--")
ax.plot(x,x+7,color="green",ls="-.")
ax.plot(x,x+8,color="green",ls=":")

#marcador
ax.plot(x,x+9,color="red",linestyle="-", lw=1, marker='+')# continua
ax.plot(x,x+10,color="red",ls="--", lw=0.5, marker='o')
ax.plot(x,x+11,color="red",ls="-.", marker='s')
ax.plot(x,x+12,color="red",ls=":", marker='1')

ax.plot(x,x+12,color="red",ls=":", marker='s', markersize=8, markerfacecolor="yellow", markeredgewidth=2, markeredgecolor='green')
```
