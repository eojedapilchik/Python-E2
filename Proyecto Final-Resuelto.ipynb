# 3: Analisis Financiero Basico

Porcentaje de Variacion Diario. 

$ r_t = \frac{p_t}{p_{t-1}} -1$

Pt es el precio en un dia x, y Pt-1 es el precio del dia anterior (x-1)
Esto indica el porcentaje de ganancia o perdida entre un dia y otro.  
Es util para evaluar volatilidad

```python
tsla['retorno']=((tsla.Close/tsla.Close.shift(1))-1)*100
```

```python
tsla.head()
```

```python
tsla['retorno2']= tsla.Close.pct_change(1)*100
```

## Retornos acumulados diarios

Con los retornos acumulados diarios vamos a intentar responder, Si hubiese invertido 1$ en la compaÃ±ia al principio de la serie de tiempo, cuanto seria su valor hoy dia

    Date                        Price
    01/01/2018                   10
    01/02/2018                   15
    01/03/2018                   20
    01/04/2018                   25

    Date                         Daily Return                  %Daily Return
    01/01/2018                 10/10 =  1                          -   
    01/02/2018                 15/10 =  3/2                       50%
    01/03/2018                 20/15 =  4/3                       33%
    01/04/2018                 25/20 =  5/4                       20%

    Date                       Cumulative Return         %Cumulative Return
    01/01/2018                  10/10 =  1                         100 %   
    01/02/2018                  15/10 =  3/2                       150 %
    01/03/2018                  20/10 =  2                         200 %
    01/04/2018                  25/10 =  5/2                       250 %

La formula del retorno acumulado es:

$ i_i = (1+r_t) * i_{t-1} $


    df[daily_cumulative_return] = ( 1 + df[pct_daily_return] ).cumprod()

```python
tsla['Retorno acumulado']=(1+(tsla.retorno/100)).cumprod()
```

```python
tsla.head()
```

```python
gm['Retorno acumulado']=(1+(gm.retorno/100)).cumprod()
```

```python
gm.tail()
```

```python
ford['Retorno acumulado']=(1+(ford.retorno/100)).cumprod()
ford.head()
```

```python
# a graficarlo

tsla['Retorno acumulado'].plot(label="Tesla", figsize=(10,8))
gm['Retorno acumulado'].plot(label="GM")
ford['Retorno acumulado'].plot(label="Ford")
plt.legend()
```
